```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['Date', 'Link', 'Header', 'Article'])
df
```

```python
# final script to scrape kanye west stories from US Weekly Website
# first for loop has range of page numbers. They are probabaly in the range of 100's 
for page_num in range(1,100):
   page_url = "https://www.usmagazine.com/wp-json/wp/v2/posts?filter[celebrity]=kanye-west&per_page=20&page="+ str(page_num)
   response = json.loads(requests.get(page_url).text)
   for article_item in response:
     link = article_item['link']
     link_response = requests.get(link)
     html_content = link_response.text

     try:
       soup = BeautifulSoup(html_content, "html.parser")
       try:
         div = soup.find_all("div", {'class': 'article-content' })
         paragraphs = div[0].find_all('p')
         text = []
         for paragraph in paragraphs:
           paragraph_text = paragraph.text
           text.append(paragraph_text)
       except:
         text = ''
       try:
         time = soup.find_all("time", {'class':"content-published-date"})
         date = time[0].text
       except:
         time = ''
       try:
         header = soup.find_all( "h1" ,{'class':"content-title"})
         header = header[0].text
       except:
         header = ''
      #  new_data = {'Date' : date, 'Link' : link, 'Header': header, 'Article' : text}
       nd = [date, link, header, text]
       df.loc[len(df.index)] = nd
     except:
       pass
  #  break
```

```python
# final script to scrape rock stories from US magazine
for page_num in range(51,101):
   page_url = "https://www.usmagazine.com/wp-json/wp/v2/posts?filter[celebrity]=dwayne-johnson&per_page=20&page="+ str(page_num)
   response = json.loads(requests.get(page_url).text)
   for article_item in response:
     try:
       link = article_item['link']
       link_response = requests.get(link)
       html_content = link_response.text
       try:
          soup = BeautifulSoup(html_content, "html.parser")
          div = soup.find_all("div", {'class': 'article-content' })
          time = soup.find_all("time", {'class':"content-published-date"})
          date = time[0].text
          paragraphs = div[0].find_all('p')
          text = ''
          for paragraph in paragraphs:
            paragraph_text = paragraph.text
            text += paragraph_text
          df = df.append({'Date' : date, 'Link' : link, 'Article' : text}, ignore_index = True)
       except:
          df = df.append({'Date' : date, 'Link' : link, 'Article' : 'null'}, ignore_index = True)
     except:
      pass
```
